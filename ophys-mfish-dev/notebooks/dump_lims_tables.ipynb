```python
from pathlib import Path
import json
import pandas
from datetime import datetime, date


def json_serial(obj):
    """JSON serializer for objects not serializable by default json code"""

    if isinstance(obj, (datetime, date)):
        return obj.isoformat()
    raise TypeError ("Type %s not serializable" % type(obj))


def read_json_table(folder, table_key):
    folder = Path(folder)
    with open(folder / f'{table_key}.json') as json_file:
        table = json.load(json_file)

    # pandas
    table = pandas.DataFrame(table)
    return table


def dump_all_ophys_LIMS_tables(c, output_folder):
    output_folder = Path(output_folder)

    #table_keys = ["specimens", "ophys_experiments", "ophys_sessions", "behavior_sessions", "visual_behavior_experiment_containers", "isi_experiments", "visual_behavior_supercontainers","cell_rois"]
    #table_keys = ["ophys_experiments", "ophys_sessions", "visual_behavior_experiment_containers"]
    table_keys = ["specimens"]

    for table_key in table_keys:
        print(table_key)
        c.execute(f"SELECT * FROM {table_key}")
        table = c.fetchall()

        with open(output_folder / f'{table_key}.json', 'w') as outfile:
            json.dump(table, outfile, default=json_serial)


```

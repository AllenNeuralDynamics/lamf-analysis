```python
from codeocean import CodeOcean
import os
from codeocean.data_asset import (
        DataAssetParams, 
        Source,
        ComputationSource,
        DataAssetSearchParams
        )
from codeocean.components import SearchFilter
```

```python
domain = os.getenv("CODEOCEAN_DOMAIN")
token = os.getenv("CODEOCEAN_TOKEN")
client = CodeOcean(domain=domain, token=token)
```

```python
computation_id = 'bfa0bfb1-68c1-4c1f-a8c0-38563384e6d3'

data_asset_params = DataAssetParams(
    name="conditioned-mean-responses_736963_2024-07-12",
    description="test",
    mount="conditioned-mean-responses_736963_2024-07-12",
    # mount="results",
    tags=["736963", "CTL"],
    source=Source(
        computation=ComputationSource(
            id=computation_id,
            # path="results/"
        ),
    )
)
    
data_asset = client.data_assets.create_data_asset(data_asset_params)
```

```python
data_asset_params = DataAssetSearchParams(
    offset=0,
    limit=10,
    sort_order="desc",
    sort_field="name",
    type="dataset",
    ownership="created",
    archived=False,
    favorite=False,
    query="",
    filters=[
        SearchFilter(
            key='subject_id',
            value=736963
        ),
    ]
)

data_assets = client.data_assets.search_data_assets(data_asset_params)
```
